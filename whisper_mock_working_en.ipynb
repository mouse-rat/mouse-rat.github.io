```python
!pip install git+https://github.com/openai/whisper.git

from google.colab import drive
drive.mount('/content/drive')

!wget --no-check-certificate --content-disposition https://github.com/mouse-rat/mouse-rat.github.io/raw/main/snapnap.mp3
!cp snapnap.mp3 /content/drive/MyDrive/snap.mp3

import whisper

fileName = "/content/drive/MyDrive/snap.mp3"
lang = "en"
model = whisper.load_model("medium")

audio = whisper.load_audio(fileName)
audio = whisper.pad_or_trim(audio)

mel = whisper.log_mel_spectrogram(audio).to(model.device)

options = whisper.DecodingOptions(language=lang, without_timestamps=True)
result = whisper.decode(model, mel, options)

print()
print()
print("-"*40)
print()
print(result.text)
print()
print("-"*40)
```
